```python
def convert_time_to_relative_numeric(
    df: pd.DataFrame,
    timeColumnTitle: str = 'origintime',
    time_unit: str = "s" # "s" seconds, "m" minutes, "h" hours, "D" days
    ):
    
    
    t0 = df[timeColumnTitle].min()
    dt = (df[timeColumnTitle] - t0) / np.timedelta64(1, time_unit)  # float
    df["t_rel"] = dt.astype(float)
    
    return df

```

```python
def preprocess_time_in_df(
    df: pd.DataFrame,
    time_col: str = "origintime",
    parse_time: bool = True,
    utc: bool = True,
    required_cols: Optional[List[str]] = None,
    dropna: bool = True,
    make_relative_time: bool = True,
    relative_time_col: str = "t_rel",
    time_unit: str = "D",   # "s", "m", "h", "D"
    ) -> pd.DataFrame:
    """
    Preprocess a earthquake catalog dataframe.

    Parameters
    ----------
    df : pd.DataFrame
        Input dataframe.
    time_col : str
        Name of the time column.
    parse_time : bool
        Whether to parse the time column as datetime.
    utc : bool
        Whether to parse datetimes in UTC.
    required_cols : list[str] or None
        Columns required to remain in the dataset.
    dropna : bool
        Whether to drop rows with missing values in required_cols.
    make_relative_time : bool
        Whether to create a relative time column.
    relative_time_col : str
        Name of the relative time column to create.
    time_unit : str
        Unit for relative time: "s", "m", "h", or "D".

    Returns
    -------
    pd.DataFrame
        Cleaned dataframe.
    """
    
    d = df.copy()
    
    if parse_time:
        d[time_col] = pd.to_datetime(d[time_col], errors="coerce", utc=utc)

    
    if required_cols is None:
        required_cols = []


    if dropna and required_cols:
        d = d.dropna(subset=required_cols).copy()

    if d.empty:
        raise ValueError("No valid rows remain after preprocessing.")

    if make_relative_time:
            if time_col not in d.columns:
                raise ValueError(f"Time column '{time_col}' not found in dataframe.")

            t0 = d[time_col].min()
            if pd.isna(t0):
                raise ValueError(f"Could not compute relative time from column '{time_col}'.")

            d[relative_time_col] = (
                (d[time_col] - t0) / np.timedelta64(1, time_unit)
            ).astype(float)

    return d

    
    
    
    
    
#     if includeTimeAsFeature:
#         # Parse time
#         d["origintime"] = pd.to_datetime(d["origintime"], errors="coerce", utc=True)
#         d_time_converted = convert_time_to_relative_numeric(df,'origintime','s')
        
#         featuresListWithoutOriginTime = subsetOfFeaturesToInclude.remove('origintime')
#         featuresListWithRelTime = featuresListWithoutOriginTime.append("t_rel")
#         print(featuresListWithRelTime)
        
# #         X = d_time_converted[["t_rel", "longitude", "latitude", "depth", "magnitude"]].to_numpy(dtype=float)
        
#         X = d_time_converted[featuresListWithRelTime].to_numpy(dtype=float)

#     else:
#         X = d[subsetOfFeaturesToInclude].to_numpy(dtype=float)
        
#     return X
```

```python
def DBSCAN_full_pipeline(
    df: pd.DataFrame,
    eps: float = 0.7,
    #if data has noise or want to detect smaller clusters, decrease min_samples
    min_samples: int = 15,
    time_unit: str = "s",  # "s" seconds, "m" minutes, "h" hours, "D" days
    timeColumnName: str = 'origintime',
    featureList: list = None,
    
    ):
    
    if featureList is None:
        featureList = ["longitude", "latitude", "depth", "magnitude"]
    
    selected_features = featureList.copy()

    
    d = df.copy()
    
    # Parse time if time is requested as a feature
    if timeColumnName in selected_features:
        if timeColumnName not in d.columns:
            raise ValueError(f"Time column '{timeColumnName}' not found in dataframe.")


        d[timeColumnName] = pd.to_datetime(d[timeColumnName], errors="coerce", utc=True)

        # Keep only rows with all required values present
        # d = d.dropna(subset=required).copy()

        # Convert time to numeric (relative time is usually better than absolute)
        t0 = d[timeColumnName].min()
        dt = (d[timeColumnName] - t0) / np.timedelta64(1, time_unit)  # float
        d["t_rel"] = dt.astype(float)

        selected_features.remove(timeColumnName)
        selected_features.append("t_rel")
            
    
    # Check feature columns exist
    missing_cols = [col for col in selected_features if col not in d.columns]
    if missing_cols:
        raise ValueError(f"Missing feature columns: {missing_cols}")

    # Drop rows with missing values in the selected features
    d = d.dropna(subset=selected_features).copy()
    
    if d.empty:
        raise ValueError("No valid rows remain after dropping missing feature values.")

    
    X = d[selected_features].to_numpy(dtype=float)
    
    # Scale so no single feature dominates distance
    scaler = StandardScaler()
    Xs = scaler.fit_transform(X)

    model = DBSCAN(eps=eps, min_samples=min_samples, metric = 'euclidean', n_jobs=-1)
    labels = model.fit_predict(Xs)

    d["dbscan_label"] = labels
    d["is_anomaly"] = (labels == -1)
    
    print("Features used:", selected_features)
    
    num_anoms = (labels == -1).sum()
    print("Number of anomalies:", )
    print("Anomalies:", num_anoms, "out of", len(d),"total")
    print("\nTop 10 clusters by size\n")
    print(d["dbscan_label"].value_counts().head(10))
    
    
    return d, model
```

```python
result_df, dbscan_model = DBSCAN_full_pipeline(
    df,
    eps = 0.1,
    #if data has noise or want to detect smaller clusters, decrease min_samples
    min_samples = 25,
    time_unit= "s",  # "s" seconds, "m" minutes, "h" hours, "D" days
    timeColumnName= 'origintime',
    featureList= ["longitude", "latitude"],
    )
```

###### must pre-process and decide which features of df want to include in dbscan before putting df as a parameter to dbscan_anomaly_detection().

scikit learn's dbscan function is in a wrapper function for modularity and reusability reasons, helping with easier experimentation of hyperparameters.

```python
def dbscan_anomaly_detection(
    df: pd.DataFrame,
    eps: float = 0.7,
    #if data has noise or want to detect smaller clusters, decrease min_samples
    min_samples: int = 15,
    time_unit: str = "s",  # "s" seconds, "m" minutes, "h" hours, "D" days
    distance_metric: str = "euclidean"
):
    """
    Run DBSCAN for anomaly detection.
    Returns a copy of df with:
      - dbscan_label: cluster id (-1 = anomaly)
      - is_anomaly: boolean
    """
    
    X_scaled = StandardScaler().fit_transform(X)

    model = DBSCAN(eps=eps, min_samples=min_samples, metric = distance_metric, n_jobs=-1)
    labels = model.fit_predict(X_scaled)

    d["dbscan_label"] = labels
    d["is_anomaly"] = (labels == -1)
    
    return d, model
```

```python
result_df,model = dbscan_anomaly_detection(X)
```
